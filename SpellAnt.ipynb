```python
def findHexWords(wordP, hexC):
  data = {}
  index = 0
  length = len(hexC)

  print("Working......")
  pbar = tqdm(total=length)
  for i in hexC:
    hex = cleanHex(i)
    #print(hex)
    center = 0
    #random.randint(0, len(hex)-1)
    wordList = []
    found = False

    for word in wordP:
      if hex[center] not in word:
        continue
      word = word.strip()
      for letter in word:
        if letter not in hex:
          #print("Char: " + letter + " Isnt in hex.") 
          found = False
          break
        #print("Char: " + letter + " is in hex.") 
        found = True
      if(found == True):
        #print('\n')
        #print(word, hex)
        wordList.append(word)
    pbar.update(n=1) 
    if(len(wordList) < 25):
      continue 
    data[index] = {"center": hex[center], "hexChars": hex, "wordPool": wordList}
    index = index + 1


  json_data = json.dumps(data)
  with open('/content/drive/MyDrive/db.json', 'w') as f:
    f.writelines(json_data)
  print("\n DONE!")
```
